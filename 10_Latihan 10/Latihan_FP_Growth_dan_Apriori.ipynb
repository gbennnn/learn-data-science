```python
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
import pandas as pd

# Dataset dummy
data = {
    'Milk': [1, 1, 1, 0, 0],
    'Bread': [1, 1, 0, 1, 1],
    'Butter': [1, 0, 1, 1, 0],
    'Eggs': [0, 1, 1, 1, 1],
    'Cheese': [0, 0, 0, 1, 1]
}

# Membuat DataFrame
df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

```python
print("\n=== Apriori ===")
frequent_itemsets_apriori = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets_apriori['length'] = frequent_itemsets_apriori['itemsets'].apply(lambda x: len(x))
rules_apriori = association_rules(
    frequent_itemsets_apriori,
    metric="confidence",
    min_threshold=0.7,
    num_itemsets=len(frequent_itemsets_apriori)  # For older versions of mlxtend
)
```

```python
print("Frequent Itemsets (Apriori):")
print(frequent_itemsets_apriori)
print("\nAssociation Rules (Apriori):")
print(rules_apriori)
```

```python
# FP-Growth
print("\n=== FP-Growth ===")
frequent_itemsets_fpgrowth = fpgrowth(df, min_support=0.4, use_colnames=True)
frequent_itemsets_fpgrowth['length'] = frequent_itemsets_fpgrowth['itemsets'].apply(lambda x: len(x))
rules_fpgrowth = association_rules(
    frequent_itemsets_fpgrowth,
    metric="confidence",
    min_threshold=0.7,
    num_itemsets=len(frequent_itemsets_fpgrowth)  # For older versions of mlxtend
)
```

```python
print("Frequent Itemsets (FP-Growth):")
print(frequent_itemsets_fpgrowth)
print("\nAssociation Rules (FP-Growth):")
print(rules_fpgrowth)
```

Rule: {Milk} â†’ {Bread}

Support: 0.6

Confidence: 0.75

Lift: 1.25

Interpretasi :
Support: 60% of transactions include both "Milk" and "Bread."

Confidence: 75% of transactions containing "Milk" also contain "Bread."

Lift: The likelihood of buying "Bread" increases by 25% when "Milk" is purchased compared to random chance.

```python
# Visualisasi
import matplotlib.pyplot as plt
import networkx as nx
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
import pandas as pd
```

```python
# 1. Scatter Plot: Support vs. Confidence
plt.figure(figsize=(10, 6))
plt.scatter(
    rules_apriori['support'],
    rules_apriori['confidence'],
    alpha=0.7,
    c=rules_apriori['lift'],
    cmap='viridis',
    s=100
)
plt.colorbar(label='Lift')
plt.title('Support vs Confidence (Color = Lift)', fontsize=14)
plt.xlabel('Support', fontsize=12)
plt.ylabel('Confidence', fontsize=12)
plt.grid(True)
plt.show()
```

```python

# 2. Network Graph of Association Rules
# Create a directed graph
G = nx.DiGraph()

# Add edges for each rule
for _, rule in rules_apriori.iterrows():
    antecedent = ', '.join(list(rule['antecedents']))
    consequent = ', '.join(list(rule['consequents']))
    G.add_edge(antecedent, consequent, weight=rule['lift'])

# Plot the network graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, seed=42)  # Positioning for better visualization
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]  # Use lift as weight

nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{d['weight']:.2f}" for u, v, d in edges})

plt.title('Association Rules Network (Edge Weight = Lift)', fontsize=14)
plt.show()
```
