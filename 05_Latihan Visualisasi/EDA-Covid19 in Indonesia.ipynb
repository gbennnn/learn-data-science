# Import Library


```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# untuk menampilkan max 200 kolom
pd.set_option('display.max_columns', 200)

# untuk mengatur gaya plot visualisasi
plt.style.use('ggplot')
```

# Memuat dan Memeriksa Dataset

```python
file_path = 'dataset/covid_19_indonesia_time_series_all.csv'
df = pd.read_csv(file_path)
```

```python
df.shape
```

```python
df.head()
```

```python
# Pastikan kolom 'date' berbentuk datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data untuk Maret 2020, Januari 2021, dan Agustus 2021
march_2020 = df[(df['Date'] >= '2020-03-01') & (df['Date'] <= '2020-03-31')]
jan_2021 = df[(df['Date'] >= '2021-01-01') & (df['Date'] <= '2021-01-31')]
aug_2021 = df[(df['Date'] >= '2021-08-01') & (df['Date'] <= '2021-08-31')]
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='New Active Cases', data=march_2020, label='Maret 2020')
plt.title('Kasus COVID-19 Maret 2020')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='New Active Cases', data=jan_2021, label='Januari 2021', color='orange')
plt.title('Kasus COVID-19 Januari 2021')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='New Active Cases', data=aug_2021, label='Agustus 2021', color='green')
plt.title('Kasus COVID-19 Agustus 2021')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Plot Maret 2020
sns.lineplot(x='Date', y='New Active Cases', data=march_2020, label='Maret 2020')

# Plot Januari 2021
sns.lineplot(x='Date', y='New Active Cases', data=jan_2021, label='Januari 2021', color='orange')

# Plot Agustus 2021
sns.lineplot(x='Date', y='New Active Cases', data=aug_2021, label='Agustus 2021', color='green')

plt.title('Perbandingan Kasus COVID-19: Maret 2020, Januari 2021, Agustus 2021')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Plot grouped bar untuk setiap periode
sns.barplot(x='Date', y='New Active Cases', data=march_2020, label='Maret 2020', color='blue', alpha=0.6)
sns.barplot(x='Date', y='New Active Cases', data=jan_2021, label='Januari 2021', color='orange', alpha=0.6)
sns.barplot(x='Date', y='New Active Cases', data=aug_2021, label='Agustus 2021', color='green', alpha=0.6)

plt.title('Perbandingan Kasus COVID-19: Maret 2020, Januari 2021, Agustus 2021')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Gabungkan data untuk membuat bar bertumpuk
merged_data = pd.concat([march_2020.assign(Period='Maret 2020'),
                         jan_2021.assign(Period='Januari 2021'),
                         aug_2021.assign(Period='Agustus 2021')])

sns.histplot(data=merged_data, x='Date', weights='New Active Cases', hue='Period', multiple='stack', palette=['blue', 'orange', 'green'])

plt.title('Perbandingan Kasus COVID-19: Maret 2020, Januari 2021, Agustus 2021')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Gabungkan data untuk heatmap
merged_data = pd.concat([march_2020.assign(Period='Maret 2020'),
                         jan_2021.assign(Period='Januari 2021'),
                         aug_2021.assign(Period='Agustus 2021')])

# Buat pivot table untuk visualisasi heatmap
pivot_data = merged_data.pivot_table(values='New Active Cases', index='Date', columns='Period')

# Plot heatmap
sns.heatmap(pivot_data, cmap='Blues', annot=True, fmt='.0f')

plt.title('Perbandingan Kasus COVID-19 dalam Heatmap')
plt.xlabel('Periode')
plt.ylabel('Tanggal')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Area plot untuk setiap periode
plt.fill_between(march_2020['Date'], march_2020['New Active Cases'], color="blue", alpha=0.4, label='Maret 2020')
plt.fill_between(jan_2021['Date'], jan_2021['New Active Cases'], color="orange", alpha=0.4, label='Januari 2021')
plt.fill_between(aug_2021['Date'], aug_2021['New Active Cases'], color="green", alpha=0.4, label='Agustus 2021')

plt.title('Perbandingan Kasus COVID-19 dengan Area Plot')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Gabungkan data untuk boxplot
merged_data = pd.concat([march_2020.assign(Period='Maret 2020'),
                         jan_2021.assign(Period='Januari 2021'),
                         aug_2021.assign(Period='Agustus 2021')])

# Plot boxplot untuk membandingkan distribusi
sns.boxplot(x='Period', y='New Active Cases', data=merged_data)

plt.title('Distribusi Kasus COVID-19: Maret 2020, Januari 2021, Agustus 2021')
plt.xlabel('Periode')
plt.ylabel('Jumlah Kasus')
plt.show()
```
